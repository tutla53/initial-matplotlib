### Importing the `library` and `CSV File` at the certain range of data

```python
import pandas as pd
import matplotlib.pyplot as plt

step_data = 10
final_idx = 3246

df = pd.read_csv('source/Visco_80_2_231122.csv')
```

### Calculating the `average` and `standard deviation` of the data

```python
print('Average: %.2f' % df.Temperature[:final_idx:step_data].mean())
print('Standard Deviation:%.2f' % df.Temperature[:final_idx:step_data].std())
```

### Setting the plot `title`, `ylabel`, `xlabel`, `xticks`, and `grid`

```python
plt.plot(df.Date[:final_idx:step_data], df.Temperature[:final_idx:step_data],'.-')

plt.title("Temperature During Viscosity Experiment (C)")
plt.ylabel("Temperature (C)")
plt.xlabel("Date Time GMT+7")

plt.xticks(df.Date[:final_idx:360], rotation=45,ha='right')
plt.grid()
plt.show()
```
