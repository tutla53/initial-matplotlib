```python
import pandas as pd
import matplotlib.pyplot as plt

visco_80_1 = pd.read_csv('source/Visco_80_2_231122.csv')
visco_80_2 = pd.read_csv('source/Visco_80_3_241122.csv')

step_data = 10
start_idx1 = 0
start_idx2 = 74

final_idx = 2595
#3246
```

### Calculating the `average` and `standard deviation` of the data

```python
fig, (ax1,ax2) = plt.subplots(2,1, sharex=True,sharey=True)

ax1.plot(visco_80_1.Date[start_idx1:final_idx-start_idx2:step_data], visco_80_1.Temperature[start_idx1:final_idx-start_idx2:step_data], linewidth = 1.2)
ax2.plot(visco_80_2.Date[start_idx2:final_idx:step_data], visco_80_2.Temperature[start_idx2:final_idx:step_data], linewidth = 1.2)

fig.suptitle("Temperature during Measuring Viscosity of 80% Glycerol")
fig.supxlabel("Time")
fig.supylabel("Temperature (°C)")

ax2.set_xticks(visco_80_2.Date[start_idx2:final_idx:360])

ax1.grid()
ax2.grid()
```
